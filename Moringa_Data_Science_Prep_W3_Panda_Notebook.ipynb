```
import pandas as pd
import numpy as np
```

```
df1 = pd.read_csv("Telcom_dataset.csv")
df1.columns = ['PRODUCT','VALUE','DATETIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
df1.head(5)
```

```
df2 = pd.read_csv("Telcom_dataset2.csv")
df2.columns = ['PRODUCT','VALUE','DATETIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
df2.head(5)
```

```
df3 = pd.read_csv("Telcom_dataset3.csv")
df3.columns = ['PRODUCT','VALUE','DATETIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
df3.head(5)
```

```
dfm = pd.concat([df1,df2])
df = pd.concat([dfm,df3])
df.head(10)
```

```
#we are going to start cleaning the concated datasets of telcom by;
#droping the country a and b because they are useless
df.drop(['COUNTRY_A','COUNTRY_B'], axis= 1, inplace= True)
df.head(3)
```

```
#we are also going to drop the cell on site because it is not needed here 
df.drop('CELL_ON_SITE', axis= 1, inplace=True)
df.head(3)
```

```
#we will drop the missng observation if there are there
df.dropna()
#will also delete the duplicates
df.drop_duplicates()
```

```
df4 = pd.read_csv('cells_geo.csv', delimiter = ';')
df4.columns = ['Unnamed','VILLES','STATUS','LOCALISATION','DECOUPZONE','ZONENAME','LONGITUDE','LATITUDE','REGION','AREA','CELL_ID','SITE_ID']
df4.head(10)
```

```
df5 = df.merge(df4, on='SITE_ID', how='inner')
df5
```

```
#we are now going to clean the merged dataset of cells_geo and telcom datasets by;
#deleting duplicates
df5.drop_duplicates()
#drop missing observations
df5.dropna()
#drop rows where all cells in row is NA
df5.dropna(how='all')
#filling missing data with zeros
df5.fillna(0)
df5.head(10)
```

```
#data analysis
#determining the most used cities on the 3 days
df5['VALUE'].groupby(df5['VILLES']).count()
```

```
#most used city for the 3 days
df5.groupby(['VILLES','STATUS'])['VALUE'].sum().max()
```

```
#cities which were the most used during business and home hours
df5.groupby([df5["DATETIME"], "VILLES"])["VALUE"].count()
```

```
#product which is used more
df5['VILLES'].groupby(df5['PRODUCT']).count()
```
